```python
from pathlib import Path
import os
import pandas as pd

DATA_DIR = Path("/Users/jana/Documents/Github/Data_Mining_Project/dataset/archive")

image_exts = {".jpg", ".jpeg", ".png", ".webp"}

all_images = []
for root, dirs, files in os.walk(DATA_DIR):
    for f in files:
        p = Path(root) / f
        if p.suffix.lower() in image_exts:
            all_images.append(p)

print("Total images found:", len(all_images))
all_images[:10]
```

```python
rows = []

for p in all_images:
    parts = p.parts
    
    split = p.parent.parent.name
    label = p.parent.name
    
    rows.append({
        "path": str(p),
        "filename": p.name,
        "split": split,
        "label": label,
        "ext": p.suffix.lower()
    })

df_images = pd.DataFrame(rows)
print(df_images.shape)
df_images.head()
```

```python
print("Split counts:")
print(df_images["split"].value_counts(), "\n")

print("Label counts:")
print(df_images["label"].value_counts(), "\n")

print("Split x Label:")
print(pd.crosstab(df_images["split"], df_images["label"]))
```

```python
output_path = Path("/Users/jana/Documents/Github/Data_Mining_Project/dataset/processed/image_metadata.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)

df_images.to_csv(output_path, index=False)
print("Saved:", output_path)
```
